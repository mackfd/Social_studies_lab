```python
for i in range(2000,3000,1):
    url = games_list['full_url'][i]
    browser = webdriver.Firefox()
    # opening url in browser via selenium
    browser.get(url)
    # checking if we have beed redirected
    soup = BeautifulSoup(browser.page_source)
    browser.quit()
    #soup = BeautifulSoup(resp, "lxml")
    tags = soup('a')
    for tag in tags: 
        if tag.get('class') == ['ng-binding'] and 'ng-click' in tag.attrs and 'role' in tag.attrs:
            print('text: ',tag.text)
            #category.append(tag.contents[0])
            games_list['pages'][i] = int(tag.text)
```
