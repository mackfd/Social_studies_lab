## Preliminar analysis of descriptive games attributes

```python
import os
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
add_data_data = pd.read_csv(r'games_list_add_data.csv', sep = '\t', encoding='utf-8',nrows=1000)
```

```python
add_data.drop(['url','full_url','overview_url'],axis=1,inplace=True)
```

```python
add_data
```

```python
play_time = add_data['play_time'].value_counts()
```

```python
plt.figure(figsize=(12,19))
plt.barh(play_time.index, play_time.values,  alpha=0.5)
plt.yticks(play_time.index)
plt.tick_params(labelsize=9)
plt.xlabel("Count", size=15)
plt.ylabel("Play time", size=15)
plt.title("Play time popularity of top 1 000 games", size=18)
for index, value in enumerate(play_time):
    plt.text(value, index, str(value))
plt.show()
plt.tight_layout()
```

```python
player_age = add_data['player_age'].value_counts()
```

```python
plt.figure(figsize=(10,6))
plt.barh(player_age.index, player_age.values,  alpha=0.5)
plt.yticks(player_age.index)
plt.tick_params(labelsize=9)
plt.xlabel("Count", size=15)
plt.ylabel("Play age", size=15)
plt.title("Play age popularity of top 1 000 games", size=18)
for index, value in enumerate(player_age):
    plt.text(value, index, str(value))
plt.show()
plt.tight_layout()
```

```python
user_players = add_data['user_players'].value_counts()
```

```python
plt.figure(figsize=(10,8))
plt.barh(user_players.index, user_players.values,  alpha=0.5)
plt.yticks(user_players.index)
plt.tick_params(labelsize=9)
plt.xlabel("Count", size=15)
plt.ylabel("# of Players", size=15)
plt.title("# of Players popularity of top 1 000 games", size=18)
for index, value in enumerate(user_players):
    plt.text(value, index, str(value))
plt.show()
plt.tight_layout()
```
