```python
def scrape(game_number, limit_pages):
    game_number = game_number
    limit_pages = limit_pages
    # iterate over given range of games 
    for game in range(game_number[0],game_number[1]+1,1):
        # initiate/clean lists for storing data for one game
        game_name_list_full = []
        nick_name_full = []
        dates_full = []
        ratings_full = []
        locs_data_full = []
        
        # getting game's name and printing to see the progress
        game_name = games_list['name'][game]
        game_id = games_list['id'][game]
        print('Game name:', game_name)
        print('Game id:', game_id)
        # for each game we are taking N (parameter limit_pages) random pages to scrape
        pages = random.randint(games_list['pages'][game], size = limit_pages)
        
        url = games_list['full_url'][game]
    
        # here we are going through pages of ranks of a game   
        for i in pages:
            print('working on page:', i) 
            page_time_start = time.time()
            # forming final url to scrape particular page 
            final_url = url + '?pageid=' + str(i)
            #print(final_url)
            # calling function
            browser = webdriver.Firefox()
            # opening url in browser via selenium
            browser.get(final_url)
            
            # passing page source data from browser to beautiful soup
            soup = BeautifulSoup(browser.page_source)
            
            # closing webpage
            browser.quit()
            
            # temporal lists to save data within iteration 
            game_name_list = []
            nick_name = []
            dates = []
            ratings = []
            locs_data = []
            
            # filling data of temporal lists via function store()
            game_name_list, nick_name, dates, ratings, locs_data = get_data(soup, game_name)
                       
            # once we get data we adding it to the games lists 
            game_name_list_full.extend(game_name_list)
            nick_name_full.extend(nick_name)
            dates_full.extend(dates)
            ratings_full.extend(ratings)
            locs_data_full.extend(locs_data)
            
            page_time_end = time.time()
            print("--- %s seconds ---" % (page_time_end - page_time_start))
            
        create_file(game_id, game_name_list_full, nick_name_full, dates_full, ratings_full, locs_data_full, game)
```

```python
def create_file(game_id, game_name_list_full, nick_name_full, dates_full, ratings_full, locs_data_full, game):
    data = []
    data = [game_name_list_full, nick_name_full, dates_full, ratings_full, locs_data_full]
    # transforming nested lists into dataframe
    df = pd.DataFrame.from_records(data)
  
    # transpose of dataframe
    df = df.transpose()
    fine_name = str(game)+'_'+str(game_id)+'_20_pages'

    df.columns = ["Game_name","Gamer", "Date", "Rating", "Location"]
    df.to_csv(r'./%s.csv' % fine_name, sep='\t', encoding='utf-8', header='true')
    print('File saved:', fine_name)
```

```python
# this function takes soup and filters with tags and classes to get info 
# once we got info it will be passed back as the lists that will be externded and used for creation of dataframe 

def get_data(soup, game_name):
    # first part where we are retrieving rating
    tags = soup('div')
    summury = soup.find_all('li', class_='summary-item summary-rating-item ng-scope')
    # loop for collection ratings 
    ratings = []
    game_name_list = []
    for tag in tags:
        
        if str('rating-angular') in str(tag.get('class')):
            rating = tag.next_element
            rating = rating.replace('\t','').replace(' ','')
            ratings.append(rating)
            game_name_list.append(game_name)
                            
            #print(tag.next_element)
    
    # second part where we are retrieving username, data and location 
    comment_header =[]
    for part_summury in summury:
        comment_header.append(part_summury.find('div', class_='comment-header'))
    
    nick_name = []
    dates = []
    locs_data = []
    for comment in comment_header:
        nick_name.append(comment.find('div', class_='comment-header-info').find('a').get_text())
        date = comment.find('span', class_='comment-header-timestamp').find('span').get_text()
        date = date.replace('\t','').replace(' ','')
        dates.append(date)
        loc = comment.find('div', class_='comment-header-location').text
        loc = loc.replace('\t','').replace(' ','')
        locs_data.append(loc)

    return game_name_list,nick_name,dates,ratings,locs_data
```
